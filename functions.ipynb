# Types of functions:

1. recursive 
2. lamba functions/ anonymous : one linear f= lambda x: x+2
3. partial functions: some arguments value are given some not
4. monkey patching

```python
import numpy as np

# Save the original function
originalMean = np.mean

# Define a patched version
def patchedMean(a, *args, **kwargs):
    print(f"np.mean called on array: {a}")
    # Let's change behavior: if array is empty, return 0 instead of raising an error
    if len(a) == 0:
        return 0
    return originalMean(a, *args, **kwargs)

# Apply the monkey patch
np.mean = patchedMean

# Test
arr = np.array([1, 2, 3])
print(np.mean(arr))  # Logs the call, returns 2.0

emptyArr = np.array([])
print(np.mean(emptyArr))  # Logs the call, returns 0 instead of error
```
