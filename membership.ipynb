#  Membership operator
Membership operators check whether an object exists inside a container.

Operators:

in

not in

They answer: is this element a member of that collection?

```python
3 in [1, 2, 3]        # True
"a" in "apple"       # True
"x" in {"x": 1}      # True (checks keys)
| Container    | Membership complexity |
| ------------ | --------------------- |
| List / Tuple | O(n)                  |
| Set          | O(1) average          |
| Dict         | O(1) average          |
| String       | O(n)                  |
```

It does not blindly loop.

It does this:

Calls y.__contains__(x) if it exists

Else falls back to iteration

Compares using ==, not is

| Container    | Membership complexity |
| ------------ | --------------------- |
| List / Tuple | O(n)                  |
| Set          | O(1) average          |
| Dict         | O(1) average          |
| String       | O(n)                  |
