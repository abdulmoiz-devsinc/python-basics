# Difference between Compiled and Interpreted Language

- Compiled languages ⇒ first compile the whole code then executes
- Interpreted languages ⇒ compile line by line and execute

Python is **Hybrid Language** becuase: 

## step 1:
the python written code is compiled to byte code by the compiler (.pyc == __ pycache __)

## step 2:
That bytecode is then Executed by the Python Virtual Machine (PVM) Instruction by instruction At runtime. That execution phase is interpreted.

# Execution flow:




This diagram shows that Python first compiles source code into bytecode using an internal compiler. The bytecode is then executed by the Python Virtual Machine. Built-in functions are pre-compiled, C-level routines that run inside the VM and provide fast, core functionality like input, output, type conversion, and object handling.

Python knows built-in functions because they are pre-defined in the builtins module and loaded into memory when the interpreter starts, making them available in every program without import.
