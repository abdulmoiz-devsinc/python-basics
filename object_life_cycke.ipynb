## Object lifecycle:

- Creation: ( __ new __ () ⇒ creates the object , __ init __() ⇒ initialize the object)
- Reference assignment (x=A(), y=x , some sort of using it)
- Reference counting (if count=0 object dies)
- Garbage collection (a=b , b=a it is a cycle so it is detected and deleted)
- Destruction (__ del __() memort freed)
