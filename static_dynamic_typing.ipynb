# Dynamic vs Static Typing

- Dynamic typed: things are checked when executed (if there is string+int in else portion it will only give error when esle executes)
- Static typed: error is thrown when compiled

Static typing checks types before execution, while dynamic typing checks types at runtime. Python is dynamically typed because variable names are bound to objects whose types are determined and verified only when operations are executed.

| Feature             | Static typing | dynamic typing|
| ------------------- | ------------- | ------------- |
| Type checked        | Compile time  | Runtime       |
| Variable type       | Fixed         | Flexible      |
| Type errors         | Early         | When executed |
| Type declarations   | Required      | Optional      |
| Runtime flexibility | Low           | High          |
