# Namespaces:
A namespace is a container that maps names to objects and prevents name conflicts by controlling the scope in which identifiers exist.
Under the hood, itâ€™s literally a dictionary.

## Ways to create a namespace: 

### 1. Module namespace: 
Every .py file is its own namespace.

### 2. Function namespace:
created when a function is called destroyed when returned

### 3. Class namespace:
Created when the class is defined, not when instantiated.
CLASSNAME.__dict__ holds the namespace
Methods are just functions stored in that dict
Instances have their own namespaces

### 4. Built-in namespace
Created at interpreter startup.
