# Identity Operator

Identity operators check whether two names refer to the same object in memory, not whether they look the same.

- is
- is not


is vs ==


a = [1, 2, 3]
b = [1, 2, 3]

a == b    # True  (same value)
a is b    # False (different objects)


Now:

c = a
c is a    # True


Because both names reference the same object.

```python
a=[1,2,3]
b=[1,2,3]
```

```python
id(a)
```

```python
id(b)
```

```python
a==b
```

```python
a is b
```
